Project: Real Estate data Analysis

Problem Definition: Analyze synthetic real estate data to understand property characteristics.

To-Do List:

Generate synthetic real estate data.
Calculate basic statistics for property features.
Create scatter plots to visualize relationships between square footage and price.
Column Names:

square_footage: Square footage of the house
bedrooms: Number of bedrooms
location: Location of the house
price: Housing price

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate Synthetic Real Estate Data
# Number of data points
num_samples = 100

# Generate synthetic data
data = {
    'square_footage': np.random.randint(1000, 5000, num_samples),
    'bedrooms': np.random.randint(1, 6, num_samples),
    'location': np.random.choice(['Downtown', 'Suburb', 'Rural'], num_samples),
    'price': np.random.randint(100000, 1000000, num_samples)
}

# Create a DataFrame
df = pd.DataFrame(data)

# 2. Calculate Basic Statistics for Property Features
statistics = df.describe()

# Print the statistics
print(statistics)

# 3. Create Scatter Plot to Visualize the Relationship between Square Footage and Price
plt.figure(figsize=(10, 6))
plt.scatter(df['square_footage'], df['price'],  color='hotpink')
plt.title('Scatter Plot: Square Footage vs. Price')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
